```python
query = ('''SELECT party_age, party_number, count(*) as ct
    FROM parties
    WHERE at_fault
    GROUP BY party_age, party_number ORDER BY party_number, ct DESC''')
df3 = pd.read_sql_query(query, con)
df3.dropna(inplace=True)

df3
# party_number, party_type, party_sobriety, party_number_killed, cellphone_use, party_race, party_number_injured, vehicle

```

```python
# create figure and axis objects with subplots()
fig,ax = plt.subplots()
# make a plot
ax.plot(df3.party_number, df3.ct, color="red" , alpha=.4, linewidth=3)
# set x-axis labe
ax.set_xlabel("party#",fontsize=14)
# set y-axis label
ax.set_ylabel("ct",color="red",fontsize=14)
# twin object for two different y-axis on the sample plot
ax2=ax.twinx()
# make a plot with different y-axis using second axis object
ax2.plot(df3.party_number,df3.party_age ,color="blue", alpha=.6, linewidth=3)
ax2.set_ylabel("age",color="blue",fontsize=14)
plt.show()
```
